```python
import random as r
import matplotlib.pyplot as plt
import numpy as np
```

```python
def blackjack(n: int, deck: list):
    done = 0
    count = 0
    copy = deck[:]
    left = len(deck)
    cards_arr = []
    card_count = 0
    
    for j in range (n):
        while count <= 21:
            if count == 21:
                done += 1
                cards_arr.append(card_count)
                break
            elif count < 21:
                index = r.randint(0, left-1)
                count += copy[index]
                copy.pop(index)
                left = len(copy)
                card_count += 1
        card_count = 0
        count = 0
        copy = deck[:]
        left = len(deck)

    failed = [0] * (n - len(cards_arr))
        
    print("Probability of getting a blackjack with given conditions: %s" % (done/n))
    plt.hist(cards_arr + failed, bins=np.arange(0,9)-0.5, rwidth=0.5, density=True)
    plt.xlabel('Cards required to sum 21 (0 = failed tests)')
    plt.ylabel('Percentage over total')
```

```python
def three_to_go(n: int, deck: list):
    done = 0
    count = 19
    copy = deck[:]
    left = len(copy)
    cards_arr = []
    card_count = 0
    
    for h in range (n):
        copy.remove(7)
        copy.remove(7)
        copy.remove(5)
        left = len(copy)
        while len(copy) > len(deck)/2:
            index = r.randint(0, left-1)
            copy.pop(index)
            left = len(copy) 
        while count <= 21:
            if count == 21:
                done += 1
                cards_arr.append(card_count)
                break
            elif count < 21:
                index = r.randint(0, left-1)
                count += copy[index]
                copy.pop(index)
                #if count > 21:
                #   cards_arr.append(0)
                left = len(copy)
                card_count += 1
        card_count = 0        
        count = 19
        copy = deck[:]
        left = len(copy)

    failed = [0] * (n - len(cards_arr))
        
    print("Probability of getting a blackjack with given conditions: %s" % (done/n))
    plt.hist(cards_arr + failed, bins=np.arange(0,5)-0.5, rwidth=0.5, density=True)
    plt.xlabel('Cards required to sum 21 (0 = failed tests)')
    plt.ylabel('Percentage over total')
```
